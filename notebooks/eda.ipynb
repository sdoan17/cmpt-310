## K-Means Clustering For Sanity Check

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Select a subset of features
numeric_cols = [
    'Admission grade',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (pass rate)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (pass rate)'
]
X = df[numeric_cols]
X_scaled = StandardScaler().fit_transform(X)
```

```python
for k in range(2,10):
    km = KMeans(n_clusters=k , random_state=42).fit(X_scaled)
    score = silhouette_score(X_scaled, km.labels_)
    print(f"K={k} silhouette_score = {score:.3f}")
```

```python
# Fit with K=3
km3 = KMeans(n_clusters=3, random_state=42).fit(X_scaled)
df['cluster'] = km3.labels_

# Visualize cluster vs true target
sns.countplot(x='cluster', hue='Target', data=df)
plt.title("Cluster assignment vs. True Outcome")
plt.show()
```

- Key finding: One cluster (Cluster 1) was almost purely Dropouts, demonstrating our features strongly distinguish at-risk students.
- Limitation: The other two clusters both captured Graduates (with a few Enrolled students mixed in), so the algorithm couldn’t cleanly separate Enrolled vs. Graduate.
